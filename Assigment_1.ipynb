```python
from datasets import load_dataset

dataset = load_dataset(
    "ainlpml/english-hindi",
    token=userdata.get("HF_TOKEN")
)
```

```python
import pandas as pd
from pandas import DataFrame
df=DataFrame(dataset)
df.head()
```

```python
df.iloc[0:999]
```

```python
col = df.columns[0]

n = len(df) // 2

new_df = pd.DataFrame({
    "english": df[col].iloc[:n].reset_index(drop=True),
    "hindi": df[col].iloc[n:].reset_index(drop=True)
})

print(new_df.head())
```

```python
new_df.head()
```

```python
(new_df.iloc[0][1])
```

```python
new_df["english"] = new_df["english"].apply(lambda x: x["text"])
new_df["hindi"] = new_df["hindi"].apply(lambda x: x["text"])
```

```python
new_df["eng_word_count"] = new_df["english"].str.split().str.len()
new_df["hin_word_count"] = new_df["hindi"].str.split().str.len()
```

```python
filtered_df = new_df[
    new_df["eng_word_count"].between(3, 60) &
    new_df["hin_word_count"].between(3, 60)
].reset_index(drop=True)
```

```python
print(type(new_df.loc[0, "english"]))
print(new_df.loc[0, "english"])
```

```python

filtered_df = new_df[
    abs(new_df["eng_word_count"] - new_df["hin_word_count"]).between(-10,10)
].reset_index(drop=True)
```

```python
new_df.head()
new_df.shape
```

```python
new_df["eng_char_count"] = new_df["english"].str.len()
new_df["hin_char_count"] = new_df["hindi"].str.len()
```

```python
new_df["diff_word_count"] = abs(new_df["eng_word_count"]-new_df["hin_word_count"])
new_df["diff_char_count"] = abs(new_df["eng_char_count"]-new_df["hin_char_count"])
```
