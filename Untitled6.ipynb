```python
import torch
import pandas as pd
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
```

```python
MODEL_NAME = "ai4bharat/indictrans2-en-indic-dist-200M"

device = "cuda" if torch.cuda.is_available() else "cpu"

tokenizer = AutoTokenizer.from_pretrained(
    MODEL_NAME,
    trust_remote_code=True,
    token=HF_TOKEN
)

model = AutoModelForSeq2SeqLM.from_pretrained(
    MODEL_NAME,
    trust_remote_code=True,
    torch_dtype=torch.float16 if device == "cuda" else torch.float32,
    token=HF_TOKEN
).to(device)

model.eval()

print(device)
```

```python
def translate(sentences, batch_size=32):

    translations = []

    for i in range(0, len(sentences), batch_size):

        batch = sentences[i:i+batch_size]

        inputs = tokenizer(
            batch,
            src_lang="eng_Latn",
            tgt_lang="hin_Deva",
            padding=True,
            truncation=True,
            max_length=256,
            return_tensors="pt"
        ).to(device)

        with torch.no_grad():
            outputs = model.generate(
                **inputs,
                max_new_tokens=256,
                num_beams=1,
                do_sample=False
            )

        translated = tokenizer.batch_decode(
            outputs,
            skip_special_tokens=True
        )

        translations.extend(translated)

    return translations
```

```python
sample_df["predicted_hindi"] = sample_df["english"].apply(translate)
```

```python
!pip install -q sacrebleu
```

```python
from sacrebleu.metrics import BLEU, CHRF, TER
```

```python
references = sample_df["hindi"].tolist()
predictions = sample_df["predicted_hindi"].tolist()
```

```python
bleu = BLEU()
chrf = CHRF()
ter = TER()

bleu_score = bleu.corpus_score(predictions, [references])
chrf_score = chrf.corpus_score(predictions, [references])
ter_score = ter.corpus_score(predictions, [references])
```

```python
with open("translation_metrics.txt", "w", encoding="utf-8") as f:
    f.write("Machine Translation Evaluation Results\n")
    f.write("=" * 40 + "\n\n")

    f.write(f"Number of sentence pairs : {len(sample_df)}\n\n")

    f.write(f"BLEU Score : {bleu_score.score:.4f}\n")
    f.write(f"chrF Score : {chrf_score.score:.4f}\n")
    f.write(f"TER Score  : {ter_score.score:.4f}\n\n")

    f.write("Detailed Results\n")
    f.write("-" * 40 + "\n")
    f.write(str(bleu_score) + "\n")
    f.write(str(chrf_score) + "\n")
    f.write(str(ter_score) + "\n")

print("Saved as translation_metrics.txt")
```
